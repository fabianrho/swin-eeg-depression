```python
import os

import pandas as pd

results_mean = pd.read_csv("results_mean.csv", index_col=0)
results_std = pd.read_csv("results_std.csv", index_col=0)

results_mean_deprnet = results_mean.loc[results_mean.model_type == "deprnet"]
# reset idnex
results_mean_deprnet.reset_index(drop=True, inplace=True)
results_std_deprnet = results_std.loc[results_std.model_type == "deprnet"]
# reset idnex
results_std_deprnet.reset_index(drop=True, inplace=True)

results_mean_swin = results_mean.loc[results_mean.model_type == "swin"]
# reset idnex
results_mean_swin.reset_index(drop=True, inplace=True)
results_std_swin = results_std.loc[results_std.model_type == "swin"]
# reset idnex
results_std_swin.reset_index(drop=True, inplace=True)
```

```python
# iter over all the results
for i, row_mean_swin in results_mean_swin.iterrows():
    row_std_swin = results_std_swin.iloc[i]

    row_mean_deprnet = results_mean_deprnet.iloc[i]
    row_std_deprnet = results_std_deprnet.iloc[i]

    row_string = f""


    for metric in ["accuracy", "precision", "recall", "f1"]:
        row_string += f" & ${row_mean_swin[metric]:.3f} \\pm {row_std_swin[metric]:.3f}$"

    for metric in ["accuracy", "precision", "recall", "f1"]:
        row_string += f" & ${row_mean_deprnet[metric]:.3f} \\pm {row_std_deprnet[metric]:.3f}$"

    # print(f'${row_mean_swin["accuracy"]:.3f} \\pm {row_std_swin["accuracy"]:.3f} & {row_mean_swin["precision"]:.3f} \\pm {row_std_swin["precision"]:.3f}$')

    print(row_string)
    # print(f'${round(row_mean["accuracy"],3)} \\pm$')
```
